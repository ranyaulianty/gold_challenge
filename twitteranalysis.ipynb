```python
tweet1 = pd.DataFrame(df['Tweet'])
tweet1.head(5)
```

```python
#Tweet cleansing
from unidecode import unidecode
import re
```

```python
#ASCII removal
def repl_ascii(text):
    return unidecode(text)
def remove_ascii(text):
    return re.sub(r"\\x[A-Za-z0-9./]+", '', unidecode(text))
tweet1['Tweet'] = tweet1['Tweet'].apply(remove_ascii)
tweet1.head(10)
```

```python
def newline_remove(s):
    return s.strip().replace(r'\n'," ")

tweet1['Tweet'] = tweet1['Tweet'].apply(newline_remove)
tweet1.head(8)
```

USER removal dilakukan karena pada data, mention sudah berganti dengan USER sehingga lebih baik dihapus agar tidak mendominasi analisis kata

```python
#Kata 'USER' removal 
def remove_user(text):
  return re.sub('USER', '', text)

tweet1['Tweet'] = tweet1['Tweet'].apply(remove_user)
tweet1.head(10)
```

RT removal dilakukan karena pada data, mention sudah berganti dengan RT sehingga lebih baik dihapus agar tidak mendominasi analisis kata

```python
#Kata 'RT' removal 
def remove_rt(text):
  return re.sub('RT', '', text)

tweet1['Tweet'] = tweet1['Tweet'].apply(remove_rt)
tweet1.head(5)
```

```python
#Emojis removal
def remove_emojis(text):
    emoj = re.compile("["
        u"\U0001F600-\U0001F64F"  # emoticons
        u"\U0001F300-\U0001F5FF"  # symbols & pictographs
        u"\U0001F680-\U0001F6FF"  # transport & map symbols
        u"\U0001F1E0-\U0001F1FF"  # flags (iOS)
        u"\U00002500-\U00002BEF"  # chinese char
        u"\U00002702-\U000027B0"
        u"\U00002702-\U000027B0"
        u"\U000024C2-\U0001F251"
        u"\U0001f926-\U0001f937"
        u"\U00010000-\U0010ffff"
        u"\u2640-\u2642" 
        u"\u2600-\u2B55"
        u"\u200d"
        u"\u23cf"
        u"\u23e9"
        u"\u231a"
        u"\ufe0f"  # dingbats
        u"\u3030"
                      "]+", re.UNICODE)
    return re.sub(emoj, '', text)

tweet1['Tweet'] = tweet1['Tweet'].apply(remove_emojis)
tweet1.head(5)
```

```python
#Punctuation removal
def punc_remove(text):
    return re.sub(r'[^\w\d\s]', '', text)

tweet1['Tweet'] = tweet1['Tweet'].apply(punc_remove)
tweet1.head(8)
```

```python
#Mejadikan seluruh kata menjadi lowercase
tweet1 = pd.DataFrame((tweet1['Tweet']).str.lower())
tweet1.head(8)
```

```python
#Menghitung jumlah kata dalam setiap tweet/cuitan
def split_kalimat(text):
  return (len(text.split()))

tweet1['total_kata'] = tweet1['Tweet'].apply(split_kalimat)
tweet1.head(3)  
```
